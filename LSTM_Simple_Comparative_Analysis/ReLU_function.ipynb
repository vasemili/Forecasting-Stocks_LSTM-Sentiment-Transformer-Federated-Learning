```python
import numpy as np
import matplotlib.pyplot as plt;
import pandas as pd;
```

```python
def relu(x):
    return max(0.0, x)

# x = 1.0
# print('Applying Relu on (%.1f) gives %.1f' % (x, relu(x)))
# x = -10.0
# print('Applying Relu on (%.1f) gives %.1f' % (x, relu(x)))
# x = 0.0
# print('Applying Relu on (%.1f) gives %.1f' % (x, relu(x)))
# x = 15.0
# print('Applying Relu on (%.1f) gives %.1f' % (x, relu(x)))
# x = -20.0
# print('Applying Relu on (%.1f) gives %.1f' % (x, relu(x)))

x = np.arange(-10,11,1)

y = np.zeros(len(x))

for i in range(len(x)):
    y[i] = relu(x[i])

plt.plot(x,y,color='black'); plt.grid();
plt.title("ReLU", fontsize=14)
#plt.title("ReLU");
plt.legend(["R(x) = max(0.0,x)"],loc='center right',bbox_to_anchor=(0.75, -0.17) , fontsize = 14);
#plt.font
```
