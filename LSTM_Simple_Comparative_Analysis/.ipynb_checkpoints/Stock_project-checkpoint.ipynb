```python
## Feature Scaling
#df_TSLA = pd.read_csv(r'C:\Users\emili\Downloads\Machine_Learning_SP_Course\Machine_Learning_Project\TSLA.csv', index_col = 'Date')
sc = StandardScaler();
#df_TSLA = df_TSLA.drop(columns =['Date']);
#X_ft = sc.fit_transform(X_feat.values);
#X_ft = pd.DataFrame(columns = X_feat.columns, data = X_ft, index = X_feat.index);

def lstm_split(data, n_steps):
    X, y = [], [];
    for i in range(len(data) - n_steps+1):
        X.append(data[i:i + n_steps, :-1]);
        y.append(data[i + n_steps - 1, -1]);
    return np.array(X), np.array(y);
```

```python
df_TSLA
```

```python
X1, y1 = lstm_split(df_TSLA.values, n_steps = 2);

train_split = 0.8;
split_idx = int(np.ceil(len(X1)*train_split));
date_index = df_TSLA.index

X_train, X_test = X1[:split_idx], X1[split_idx:];
y_train, y_test = y1[:split_idx], y1[split_idx:];
X_train_date, X_test_date = date_index[:split_idx], date_index[split_idx:];

print(X1.shape, X_train.shape, X_test.shape, y_test.shape)
```

```python
lstm = Sequential();
lstm.add(LSTM(32, input_shape = (X_train.shape[1], X_train.shape[2])));
lstm.add(Dense(1));
lstm.compile(loss = 'mean_squared_error', optimizer = 'adam');
lstm.summary()
```

```python
history = lstm.fit(X_train, y_train, epochs = 100, batch_size = 4, verbose = 2, shuffle = False);
```

```python
y_pred = lstm.predict(X_test);
```
